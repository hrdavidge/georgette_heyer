```python
import pandas as pd
import pdb
import numpy as np
```

```python
columns = ['fav_novel1', 'fav_novel2', 'fav_novel3', 'fav_novel4', 'fav_novel5', 'least_novel1', 'least_novel2', 'least_novel3', 'least_novel4', 'least_novel5']
fav_columns = ['fav_novel1', 'fav_novel2', 'fav_novel3', 'fav_novel4', 'fav_novel5']
least_columns =['least_novel1', 'least_novel2', 'least_novel3', 'least_novel4', 'least_novel5']
year_names = ['fav_novel_year1', 'fav_novel_year2', 'fav_novel_year3', 'fav_novel_year4', 'fav_novel_year5', 'least_novel_year1', 'least_novel_year2', 'least_novel_year3', 'least_novel_year4', 'least_novel_year5']
year_names_fav = ['fav_novel_year1', 'fav_novel_year2', 'fav_novel_year3', 'fav_novel_year4', 'fav_novel_year5']
year_names_least = ['least_novel_year1', 'least_novel_year2', 'least_novel_year3', 'least_novel_year4', 'least_novel_year5']
era_number_3 = ['fav1', 'fav2', 'fav3', 'least1', 'least2', 'least3']
era_number_4 = ['fav1', 'fav2', 'fav3', 'fav4', 'least1', 'least2', 'least3', 'least4']
```

```python
# read in dataframes
survey_df = pd.read_csv('master_csv.csv')
novels_df = pd.read_csv('novels.csv')
#characters_df = pd.read_csv('characters.csv')
```

```python
# juect selecting the rows I want from the novels dataframe
novels_df = novels_df[['novel', 'pub']]

# loop through the columns
for col_num in range(0, len(columns)):
    # join novels_df onto survey_df
    survey_df = survey_df.merge(novels_df, left_on=[columns[col_num]], right_on=['novel'], how='left')
    # delete column 'novel' from survey_df
    survey_df = survey_df.drop(labels = 'novel', axis = 1)
    # rename the new pub col to year_names[col_num]
    survey_df = survey_df.rename(index=str, columns={'pub': year_names[col_num]})
```

```python
# just choose the rows from survey_df where the user has filled out all of the favourite novels
fav_survey_df = survey_df[survey_df[fav_columns[0]].notnull() & survey_df[fav_columns[1]].notnull() & survey_df[fav_columns[2]].notnull() & survey_df[fav_columns[3]].notnull() & survey_df[fav_columns[4]].notnull()]

least_survey_df = survey_df[survey_df[least_columns[0]].notnull() & survey_df[least_columns[1]].notnull() & survey_df[least_columns[2]].notnull() & survey_df[least_columns[3]].notnull() & survey_df[least_columns[4]].notnull()]
```

Looking at three eras, early, middle, late

```python
fav_survey_df_3 = fav_survey_df
# add on era columns
fav_survey_df_3['fav1'] = 0
fav_survey_df_3['fav2'] = 0
fav_survey_df_3['fav3'] = 0

# looping through each row of fav_survey_df_3 - loading favourites in
for row in range(0, len(fav_survey_df_3)):
    # loop through the 3 era columns
    for col in year_names_fav:
        if fav_survey_df_3[col].iloc[row] < 1941:
            fav_survey_df_3['fav1'].iloc[row] = fav_survey_df_3['fav1'].iloc[row] + 1
        if fav_survey_df_3[col].iloc[row] > 1933 and fav_survey_df_3[col].iloc[row] < 1957:
            fav_survey_df_3['fav2'].iloc[row] = fav_survey_df_3['fav2'].iloc[row] + 1
        if fav_survey_df_3[col].iloc[row] > 1941:
            fav_survey_df_3['fav3'].iloc[row] = fav_survey_df_3['fav3'].iloc[row] + 1  

least_survey_df_3 = least_survey_df
# add on era columns
least_survey_df_3['least1'] = 0
least_survey_df_3['least2'] = 0
least_survey_df_3['least3'] = 0

# looping through each row of survey_df_3 loading least favourites in
for row in range(0, len(least_survey_df_3)):
    # loop through the 3 era columns
    for col in year_names_least:
        if least_survey_df_3[col].iloc[row] < 1941:
            least_survey_df_3['least1'].iloc[row] = least_survey_df_3['least1'].iloc[row] + 1
        if least_survey_df_3[col].iloc[row] > 1933 and least_survey_df_3[col].iloc[row] < 1957:
            least_survey_df_3['least2'].iloc[row] = least_survey_df_3['least2'].iloc[row] + 1
        if least_survey_df_3[col].iloc[row] > 1941:
            least_survey_df_3['least3'].iloc[row] = least_survey_df_3['least3'].iloc[row] + 1  
```
